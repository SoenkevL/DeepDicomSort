```python
import os

import matplotlib.pyplot as plt
import numpy as np
import os
from PIL import Image
```

```python
image_path = '/home/soenke/Internship/ResultsFromServer/finalExperiments/exp1/no_freeze/no_binning/example_images'
```

```python
image_paths = []
for root, dirs, files in os.walk(image_path):
    if files:
        for file in files[2:4]:
            if not 'csv' in file:
                image_paths.append(os.path.join(root, file))
```

```python
n_cols = 4
n_rows = int(np.ceil(len(image_paths)/n_cols))
```

```python
len(image_paths)/n_cols
```

```python
fig, axs = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(10,40), frameon=False, gridspec_kw=dict(wspace=0.01))
for ax, path in zip(axs.flatten(), image_paths):
    img = Image.open(path)
    img = np.array(img)
    imageName = os.path.split(path)[-1]
    imageName = os.path.splitext(imageName)[0]
    imageName = imageName.split('__')
    dataset = imageName[-2].split('_')[-1]
    weighting = imageName[0].split('_')[-1]
    if weighting == 'c':
        weighting = 'T1_c'
    imageName = '_'.join((dataset, weighting))
    print(weighting, dataset)
    ax.imshow(img)
    ax.axis('off')
    ax.set_title(imageName)
plt.show()
fig.savefig(os.path.join(image_path, 'combined.png'))
```
