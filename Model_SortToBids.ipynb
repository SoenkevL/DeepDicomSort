```python
def get_out_file_name(root_out_folder,subject_label, session_label, sub_type,
                      ce_label, modality_label, run_index):
    run_label = 'run-' + str(run_index)

    scan_labels = list(filter(None,
                              [subject_label, session_label,
                               ce_label, run_label, modality_label]))

    out_file_name = '_'.join(scan_labels) + '.nii.gz'

    out_directory = os.path.join(root_out_folder, subject_label,
                                 session_label, sub_type)

    out_file = os.path.join(out_directory, out_file_name)

    return out_file, out_directory

def create_directory(dir):
    if not os.path.exists(dir):
        os.makedirs(dir)
```

```python
def getAttributes(full_dataframe, minCertainty, root_folder):
    session_dict = dict()
    BIDS_frame = pd.DataFrame()
    for idx in tqdm(range(len(full_dataframe))):
        dataframe = full_dataframe.iloc[idx,:]
        vote = dataframe['vote']
        patientID = dataframe['NIFTI_path'].split(dataframe['NIFTI_name'])[0]
        patientID = patientID.split('/')[-2].replace('-','')
        subject_label = f'sub-{patientID}'
        splitted_elements = dataframe['NIFTI_name'].split('__')
        session_ID = splitted_elements[2]
        cur_session_index = len(session_dict)+1
        if session_ID not in session_dict:
            session_dict[session_ID] = cur_session_index
            session_num = cur_session_index
        else:
            session_num = session_dict[session_ID]
        session_label = 'ses-' + str(session_num)
        #this part is specific for a certain label map
        if vote == 0:
            sub_type = 'anat'
            modality_label = 'T1w'
        elif vote == 1:
            sub_type = 'anat'
            modality_label = 'T2w'
        elif vote == 3:
            sub_type = 'anat'
            modality_label = 'PD'
        elif vote == 2:
            sub_type = 'anat'
            modality_label = 'FLAIR'
        elif vote == 4:
            sub_type = None
            modality_label = None
        else:
            sub_type = None
            modality_label = None
        ce_label = None
        current_run_index = 1
        out_file = ''
        out_directory = ''
        if dataframe['certainty'] >= minCertainty and modality_label: #
            out_file, out_directory = get_out_file_name(root_folder,subject_label, session_label,
                                                        sub_type, ce_label, modality_label,
                                                        current_run_index)
    
            while os.path.exists(out_file):
                current_run_index += 1
                out_file, out_directory = get_out_file_name(root_folder,subject_label, session_label,
                                                            sub_type, ce_label, modality_label,
                                                            current_run_index)
            if not os.path.exists(out_directory):
                os.makedirs(out_directory)

            #shutil.copy(dataframe['NIFTI_path'], out_file)
    
        temp = pd.DataFrame([[dataframe['NIFTI_name'],subject_label, ce_label, modality_label, current_run_index, session_label, sub_type, out_file, out_directory]], columns=['NIFTI_name','subject_label', 'ce_label', 'modality_label', 'run_label', 'session_label', 'sub_type','out_file','out_directory'])
        
        if BIDS_frame.empty:
            BIDS_frame = temp
        else:
            BIDS_frame = pd.concat([BIDS_frame,temp])
    return BIDS_frame
```

```python
bids = getAttributes(Results_short, moveWithMinCertainty, RootFolder)
```
